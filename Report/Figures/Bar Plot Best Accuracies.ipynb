```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
from collections import namedtuple
#from matplotlib.pyplot import figure

#This sets the font of all items to the font specified by the kwargs object, font
font = {'family' : 'roman',
        'weight' : 'bold',
        'size'   : 17}

matplotlib.rc('font', **font)

# SVM, RF, MLP, CNN, RNN
n_groups = 5

#means_men = (20, 35, 30, 35, 27)
means_nmel = (94.39, 92.29, 92.05, 93.69, 87.00)
max_nmel = max(means_nmel)
means_mel_normalized = [x / max_nmel for x in means_nmel]
#std_men = (2, 3, 4, 1, 2)

means_all_features = (94.39, 93.45, 92.52, 92.75, 89.01)
max_all_features = max(means_all_features)
means_all_features_normalized = [y / max_all_features for y in means_all_features]
#std_women = (3, 5, 2, 3, 3)

fig, ax = plt.subplots()

fig = plt.gcf()
fig.set_size_inches(13, 9)
#fig.savefig('test2png.png', dpi=100)

index = np.arange(n_groups)
bar_width = 0.35

opacity = 0.8
error_config = {'ecolor': '0.8'}

rects1 = ax.bar(index, means_nmel, bar_width,
                alpha=opacity, color='b',
                error_kw=error_config,
                label='20 MFCCs')

rects2 = ax.bar(index + bar_width, means_all_features, bar_width,
                alpha=opacity, color='r',
                error_kw=error_config,
                label='Nmel+ZCR+RMSE+SF')

ax.set_ylim([80,95])
ax.set_xlabel('Model')
ax.set_ylabel('Accuracy %')
ax.set_title('Accuracies for each model')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(('SVM', 'RF', 'MLP', 'CNN', 'RNN'))
ax.set_yticks(np.arange(80, 96, 1.1))
ax.legend()

fig.tight_layout()
plt.grid(axis='y', linestyle='--')
plt.show()

fig.savefig('test2.png', dpi=900)
```
