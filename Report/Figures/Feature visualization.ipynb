### Feature visualization of individual audio segments using Librosa 

```python
import matplotlib.pyplot as plt
import numpy as np
import librosa.display
import matplotlib
```

```python
#Visualize an STFT power spectrum

#This sets the font of all items to the font specified by the kwargs object, font
font = {'family' : 'roman',
        'weight' : 'bold',
        'size'   : 18}

matplotlib.rc('font', **font)

y , sr = librosa.load('Avalinguo - Alan and Eduardo segment 6 - A.mp3', sr=None) # Using my own audio segment
#y, sr = librosa.load(librosa.util.example_audio_file())

plt.figure(figsize=(20, 20))

no_images = 6

# Plot of linear-frequency-power spectrum
D = librosa.amplitude_to_db(librosa.stft(y), ref=np.max) #Convert amplitude into db
plt.subplot(no_images, 2, 1)
librosa.display.specshow(D, y_axis='linear') #Linear Scale
plt.colorbar(format='%+2.0f dB')
plt.title('Linear-frequency power spectrogram')

# The above plot on a logarithmic scale
plt.subplot(no_images, 2, 2)
#plt.figure(figsize=(20, 10))
librosa.display.specshow(D, y_axis='log') #Log scale
plt.colorbar(format='%+2.0f dB')
plt.title('Log-frequency power spectrogram')

# MFCCs
C = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=20)
plt.subplot(no_images, 2, 3)
librosa.display.specshow(C, y_axis='linear')
plt.colorbar(format='%+2.0f dB')
plt.title('MFCC')

# Draw RMSE spectrogram
E = librosa.feature.rmse(y=y)
plt.subplot(no_images, 2, 4)
librosa.display.specshow(E)
plt.colorbar(format='%+2.0f dB')
plt.title('RMSE')

# ZCR
F = librosa.feature.zero_crossing_rate(y=y)
plt.subplot(no_images, 2, 5)
librosa.display.specshow(F, x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.title('ZCR')

# Chroma
G = librosa.feature.chroma_cqt(y=y, sr=sr)
plt.subplot(no_images, 2, 6)
librosa.display.specshow(G, y_axis='chroma', x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.title('Chromagram')

#Save all subplots in one image
#plt.savefig('Feature visualization.png', dpi=200)
plt.savefig('Low level.png', dpi=200)

# Spectral Flux
#G = librosa.onset.onset_strength(y=y,sr=sr)
#plt.subplot(no_images, 2, 6)
#librosa.display.specshow(G)
#plt.colorbar(format='%+2.0f dB')
#plt.title('SF')


#y, sr = librosa.load(y, duration=10.0)
#G = librosa.stft(y)
#times = librosa.frames_to_time(np.arange(D.shape[1]))
#plt.figure()
#librosa.display.specshow(librosa.amplitude_to_db(G, ref=np.max), y_axis='log', x_axis='time')
#plt.title('Power spectrogram')
```

### Save files for git gif 

```python
import matplotlib.pyplot as plt
import numpy as np
import librosa.display
import matplotlib

y , sr = librosa.load('Avalinguo - Victor and Abraham segment 123.mp3', sr=None) # Using my own audio segment

D = librosa.amplitude_to_db(librosa.stft(y), ref=np.max) #Convert amplitude into db

plt.figure(figsize=(8, 5))

plt.subplot(1, 1, 1)
librosa.display.specshow(D,y_axis='log')

plt.title('Log Power spectrogram')

plt.ylabel('Hz')
plt.title('Low Fluency Level')
plt.savefig('Low level.png', dpi=200)
```
