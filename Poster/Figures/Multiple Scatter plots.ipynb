### Code for scatter plots of accuracies with increasing MFFCs

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

# Set font size
font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 30}

matplotlib.rc('font', **font)
# Load data
# (n_mel, accuracy)
svm = ([5, 10, 12, 20], [86, 89.49, 92.06, 94.39])
rf = ([5, 10, 12, 20],[84.80, 89, 90.42, 92.29])
mlp = ([5, 10, 12, 20],[78, 88.78, 89.01, 92.05])
cnn = ([5, 10, 12, 20],[80, 85.04, 87.61, 93.69])
rnn = ([5, 10, 12, 20],[78.90, 85.04, 86.44, 87])

data = (svm, rf, mlp, cnn, rnn)
colors = ("red", "green", "blue", "cyan", "magenta")
groups = ("SVM", "RF", "MLP", "CNN", "RNN") 
 
# Create plot
fig = plt.figure(figsize=(15, 9), dpi=100)
plt.xlabel("Number of MFCCs")
plt.ylabel("Accuracy %")
ax = fig.add_subplot(1, 1, 1)
ax.grid(color='k', linestyle='-', linewidth=0.5)

for data, color, group in zip(data, colors, groups):
    x, y = data
    ax.scatter(x, y, alpha=1, c=color, edgecolors='none', s=80, label=group)
    ax.plot(x, y, c=color, linewidth=4)
 
plt.title('Accuracy performance for increasing MFCCs')
plt.legend(loc=2)
plt.savefig("Accuracies MFCCs.png", dpi=300)
plt.show()
```
